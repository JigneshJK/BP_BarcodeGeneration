### **Problem Statement:**

Barcode generation is one of the most common task while automating document based processes. There are various kinds of barcode which business might need to replace number with barcode image


Blueprism do not provide any utility to generate such barcodes, business might be using alternate ways like free websites 
(Eg:[Free Online Barcode Generator](https://barcode.tec-it.com/en/UPCA?data=72527273070)) for their manual processes but when it comes to automation SLIM dose not support such sources due to unavailability of vendor contract

Such scenarios might cause loss of usecase just due to unavailability of way to generate barcode

Another option is to use MS Excel fonts to generate barcode easily but most of them as paid ones & requires additional cost to process, plus acquiring such fonts might take different process & time


### **Solution:**

Best solution is to write inhouse code to cater this requirement

Python provides lots of prebuild libraries which we can utilize for such cases, since python is opensource platform - No separate SLIM approval is required

We have used python-barcode to generate barcodes in python & pillow to save it as image.

Formats supported by this library are :

**Code 39, Code 128, PZN7, EAN-13, EAN-8, JAN, ISBN-13, ISBN-10, ISSN, UPC-A, EAN14, GS1-128**

[㊙️ python-barcode — python-barcode 0.14.0 documentation](https://python-barcode.readthedocs.io/en/stable/)


To integrate it with Blueprism used object Utility_Object_Generic_Python

Refer KB  - **KB0310275** on how to configure Python in VM to use this VBO

Reusable **Utility_Object_Generic_Barcode** is already created for 2 sample formats EAN13 & UPCA, this can be easily extended to any other supported formats by minor changes in the code

### **Python Codes:**

```python
#1) UPCA Barcdoe 

# import UPCA from barcode module
from barcode import UPCA

# import ImageWriter to generate an image file
from barcode.writer import ImageWriter

# Make sure to pass the number as string
number = '72527273070'

# Now, let's create an object of EAN13 class and
# pass the number with the ImageWriter() as the
# writer
my_code = UPCA(number, writer=ImageWriter())

# Our barcode is ready. Let's save it.
my_code.save("C:/Users/Jignesh.Khakhriya/OneDrive - Shell/Documents/UPCA_Code")


#------------------------------------------------------------------------------

#1) EAN13 Barcdoe 

# import EAN13 from barcode module
from barcode import EAN13

# import ImageWriter to generate an image file
from barcode.writer import ImageWriter

# Make sure to pass the number as string
number = '5011987027121'

# Now, let's create an object of EAN13 class and
# pass the number with the ImageWriter() as the
# writer
my_code = EAN13(number, writer=ImageWriter())

# Our barcode is ready. Let's save it.
my_code.save("UPCA_Code")
```
